Em um cercado existem vários patos e coelhos. Escreva uma função em linguagem 
Python que recebe como parâmetro o total de cabeças e o total de pés, e retorne dois 
valores: o total de patos e o total de coelhos que se encontram nesse cercado. (4,0 
pontos)

Teste essa função elaborando um script que vai receber do usuário dois valores. Para 
tanto, leve em consideração que os dois últimos dígitos de seu RA sejam iguais a Y.


```python
def quant(cabecas, patas):
    patos = 0
    coelhos = 0 

    coelhos = (patas - (2 * cabecas))/2
    patos = cabecas - coelhos

    print(f"Patos: {patos}")
    print(f"Coelhos: {coelhos}")


c = int(input("Insira o número de cabeças: "))
p = int(input("Insira o número de patas: "))

quant(c, p)
```
