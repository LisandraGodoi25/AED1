Exercício 1: Verificar se o número é divisível por 5 e 3 ao mesmo tempo:

```python
num = int(input("Insira um número inteiro: "))
#print(f"Num: {num}")

aux3 = num
aux5 = num

while aux3 > 0:
    aux3 = aux3 - 3
    #print(f"Novo num: {aux3}")


while aux5 > 0:
    aux5 = aux5 - 5
    #print(f"Novo num: {aux5}")


if aux3 == 0 and aux5 == 0:
    print(f"O número {num} é divisível por 3 e 5")
else:
    print(f"O número {num} NÃO é divisível por 3 e 5")
```

Exercício 2: Sistema de descontos:
|Valor|Desconto|
|----|------|
|Até 1000|10%|
|De 1001 a 5000|20%|
|Acima de 5000|30%|

```python
compra = float(input("Insira o valor da compra: "))

if compra <= 1000:
    nominal = "10%"
    desconto = (compra / 100) * 10
elif compra >= 1001 and compra <= 5000:
    nominal = "20%"
    desconto = (compra / 100) * 20
elif compra > 5000:
    nominal = "30%"
    desconto = (compra / 100) * 30

print(f"Com a compra de R${compra}, seu desconto nominal é de {nominal}, equivalendo a {desconto}")
```

Exercício 3: Cálculo de latas de tinta.
Entradas: LARGURA, COMPRIMENTO, QUANTIDADE L DE TINTA

Cada Litro de tinta pinta 3 metros

Sobre o aposento: A altura do aposento mede 2,8m, deverão ser pintadas apenas as paredes. O aposento tem apenas uma porta medindo 0,8m x 2,1m.

```python
largura = float(input("Insira a largura do aposento em metros: "))

comprimento = float(input("Insira o comprimento do cômodo em metros: "))

altura = 2.8

porta = 0.8 * 2.1

areaparedes = (comprimento * altura * 2) + (largura * altura * 2) - porta

print(areaparedes)

tinta = (areaparedes / 3)

print(f"Será necessário aproximadamente {round(tinta)} baldes de 1L de tinta")
```
