```python
def conversao(celcius):
    f = celcius * (9/5) + 32
    return f
```

Cria funcao que ve se é par ou não

```python
def checknum(num):
    if num % 2 == 0:
        even = True
    else:
        even = False

    return even 
```

```python
checknum(2345)
```

```python
checknum(2345678876)
```

```python
number = int(input("Insira um número: "))

if checknum(number):
    print("O número é par!")
else:
    print("O número é ímpar!")
```

Faça o cálculo do IMC

```python
peso = float(input("Insira o peso em Kg: "))
altura = float(input("Insira a altura em metros: "))
```

```python
def calcular_imc(peso,altura):
    imc = peso / (altura*altura)
    return imc
```

```python
def classificar_imc(imc):
    if imc < 18.5:
        print("Abaixo do peso")
    elif imc <= 24.9:
        print("Peso normal")
    elif imc <= 29.9:
        print("Sobrepeso")
    else:
        print("Obesidade")
```

```python
classificar_imc(calcular_imc(peso, altura))
```

Descobrir se um número é primo

```python
number = int(input("Insira um número: "))
primo = True
aux = 2


while True:
    ##print(f"{aux}")

    if number <= 1:
        primo = False
        break
    
    if aux == (number):
        break

    if number % aux == 0:
        ##print(f"{number} % {aux} = {number % aux}")
        primo = False
        break


    aux += 1



print(primo)
```
