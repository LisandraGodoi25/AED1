# Fazer um jogo de pedra, papel e tesoura

```python
from random import choice
from time import sleep

jokenpo = ['PEDRA', 'PAPEL', 'TESOURA']

print("Vamos jogar Jo Ken Po? ")
sleep(0.5)
print("As regras são simples...")
sleep(0.5)
print("PEDRA AMASSA TESOURA")
sleep(0.5)
print("TESOURA CORTA PAPEL")
sleep(0.5)
print("PAPEL EMBRULHA PEDRA")

sleep(1)
humancount = 0
computercount = 0
empatecount = 0
human = ""
computer = ""
continueq = ""

def count():
    print("Vamos lá? No JÁ, mostre a sua mão!")
    sleep(0.25)
    print("3...")
    sleep(0.5)
    print("2...")
    sleep(0.75)
    print("1...")
    sleep(1)
    print("JÁ!!!")

def whowins(human, computer):

    global empatecount, computercount, humancount

    if human == computer:
        print("Empate!! Não valeu, vamos de novo...")
        empatecount += 1
    elif (human == 'TESOURA' and computer == 'PAPEL') or (human == 'PAPEL' and computer == 'PEDRA') or (human == 'PEDRA' and computer == 'TESOURA'):
        print("Você venceu dessa vez! Deixe-me me vingar")
        humancount += 1
    else:
        print("Eu venci! Eu venci! Hehehe Vamos de novo?")
        computercount += 1

def countvictories(empate, human, computer):
    print(f'''
    ===PLACAR====
    Empates: {empate}
    Suas vitórias: {human}
    Vitórias do computador: {computer}''')

while True:

    if continueq:
        break

    count()

    computer = choice(jokenpo)
    #print(f"Computador: {computer}")
    human = input().upper()
    #print(f"Humano: {human}")
    
    print(f"C:{computer} X H:{human}")
    sleep(0.25)

    whowins(human, computer)

    countvictories(empatecount, humancount, computercount)

    sleep(0.5)
    continueq = input("Vamos de novo? ").upper() == 'CANSEI'


```
