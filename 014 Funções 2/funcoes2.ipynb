Função que calcula o valor a ser combrado pelos serviços de entrega de uma loja virtual.

O custo é calculado tomando como base:

- O peso do produto:
    |PESO|PREÇO|
    |---|---|
    |Até 500g|R$15,00|
    |Até 2kg|R$43,00|
    |Acima de 2kg|R$103,00|

- O Estado de destino:
    |ESTADO|PREÇO|
    |---|---|
    |Norte e Nordeste|R$50,00|
    |Centro Oeste e Sul|R$25,00|
    |Sudeste|R$0,00|

- Presente (se for presente)
    Considere sempre que não é presente
    presente = False

    Se for presente, acrescentar R$10,00
    

```python
N = [  # Norte
    "AC", "AP", "AM", "PA", "RO", "RR", "TO"]

NE = [  # Nordeste
    "AL", "BA", "CE", "MA", "PB", "PE", "PI", "RN", "SE"]

CO = [  # Centro-Oeste
    "DF", "GO", "MT", "MS"]

SE = [  # Sudeste
    "ES", "MG", "RJ", "SP"]

S = [  # Sul
    "PR", "RS", "SC"]
```

```python
#Essa função carrega o frete de um envio da Amazon
print("Bem-vindo a Amazon! Para calcular seu frete, insira as informações a seguir:")

#As variárveis do frete são peso em KG, usar .2
peso_input = float(input("Insira o peso do pacote em KG: "))
#, estado me SIGLA de duas letras
estado_input = input("Insira o estado(SIGLA) para qual será enviado: ").upper()
#True/False para presente
presente_input = input("O envio será para presente? [S/N] ").lower() == "s"



print(f"Peso: {peso}")
print(f"Região: {estado}")
print(f"E estado de presente {presente}")

#Função secundária utilizada dentro da frete_func
def peso_func(peso):
    #print(f"Recebi o peso: {peso}")
    p = 0
    if peso <= 0.5:
        p = 15.0
    elif peso <= 2:
        p = 43.0
    elif peso > 2:
        p = 103.0

    #print(f"Com o peso {peso}, o preço adicionou {p}")
    return p

#Função secundária utilizada dentro da frete_func
def estado_func(estado):
    e = 0
    if estado in N or estado in NE:
        e = 50.0
    elif estado in S or estado in CO:
        e = 25.0
    elif estado in SE:
        e = 0

    #print(f"Com o estado {estado}, o preço adicionou {e}")
    return e

#Função principal que soma os valores adicionados pelo peso e estado, e acrescenta 10 se for presente
def frete_func(peso,estado,presente=False):
    p = peso_func(peso)
    e = estado_func(estado)
    total = 0

    if presente:
        #print(f"{p} + {e} + 10")
        total = 10 + p + e
    else:
        #print(f"{p} + {e}")
        total = p + e
    
    return(total)


#Guarda o valor final da função
total = frete_func(peso_input,estado_input,presente_input)

#Imprime o valor final da função
print(f"\nO frente total é de: {total}")
```
