Receba um numero ilimitado de nomes de cidades e SAIR para quebrar

Ordenar em ordem crescente

Mostrar uma cidade por linha



```python
cidades = []

while True:
    
    city = input("Insira o nome de uma cidade e digite SAIR para parar: ").upper()

    if city == "SAIR":
        break

    cidades.append(city)
    
#print(f"cidades: {cidades}")

cidades.sort()

print("Lista de cidades: ")
for city in range(len(cidades)):
    print(cidades[city].capitalize())
```

Simule 6000 jogadas de um dado de 6 faces
frequencia de cada face

```python
from random import randint

aux1 = 0
aux2 = 0
aux3 = 0
aux4 = 0
aux5 = 0
aux6 = 0

for vez in range(6000):
    dado = randint(1,6)
    #print(dado)
    if dado == 1:
        aux1 += 1

    if dado == 2:
        aux2 += 1

    if dado == 3:
        aux3 += 1
    
    if dado == 4:
        aux4 += 1

    if dado == 5:
        aux5 += 1

    if dado == 6:
        aux6 += 1


print(f"1: {aux1} vezes")
print(f"2: {aux2} vezes")
print(f"3: {aux3} vezes")
print(f"4: {aux4} vezes")
print(f"5: {aux5} vezes")
print(f"6: {aux6} vezes")

print(aux1+aux2+aux3+aux4+aux5+aux6)
```
