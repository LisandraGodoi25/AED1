```python
a = [0,1,2,3,4]

b = [5,6,7,8,9]

result = []
```

```python
for num in range(len(a)):

    #print(num)
    result.append(a[num] + b[num])
    print(f"{a[num]} + {b[num]} = {result[num]}")


print(result)
```
